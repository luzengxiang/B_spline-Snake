```python
import BSpline_Snake.GVF as GVF
import BSpline_Snake.B_spline as B_spline
import cv2
from PIL import Image
import numpy as np
import BSpline_Snake.Snake as Snake
image = cv2.imread("test2.png")


gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
gray_image = np.rot90(gray_image)
gray_image = cv2.resize(gray_image,(200,200))

image_GVF = GVF.GVF_generator(gaussian_size=3,mu=0.8).from_gray_image(gray_image,[200,200])
image_GVF.plot()
```

```python
spline = Snake.Contour_Extractor().fit_B_spline(image_GVF,image=gray_image)
```

```python
spline.plot_spline()
image_GVF.plot()
```

```python
result_matrix_x, result_matrix_y =spline.get_sample_force(image_GVF)
force = result_matrix_x ** 2 + result_matrix_y ** 2
sample_index,knot_index = np.unravel_index(np.argmax(force, axis=None), force.shape)
spline.add_knot(knot_index,sample_index)
spline.plot_spline()
```
