```python
image = cv2.imread("test.jpg")
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
image_resized = cv2.resize(gray_image,dsize=(100,100))
image_resized = image_resized / 255
```

```python
sobelx = cv2.Sobel(image_resized,cv2.CV_64F,1,0,ksize=sobel_size)
sobely = cv2.Sobel(image_resized,cv2.CV_64F,0,1,ksize=sobel_size)
gradient_magnitude = sqrt(sobelx*sobelx+sobely*sobely)

sobelx = cv2.Sobel(gradient_magnitude,cv2.CV_64F,1,0,ksize=sobel_size*2-1)
sobely = cv2.Sobel(gradient_magnitude,cv2.CV_64F,0,1,ksize=sobel_size*2-1)
sobelx = sobelx / sobelx.max()
sobely = sobely / sobely.max()

gradient_magnitude2 = sobelx*sobelx+sobely*sobely
```
